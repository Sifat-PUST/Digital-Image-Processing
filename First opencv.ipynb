```python
# Edge detection kernel (Sobel X filter)
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
filtered_image = np.zeros_like(gray_img)  # Same size as the input image


# Get image dimensions
img_height, img_width = gray_img.shape
kernel_size = kernel.shape[0]
offset = kernel_size // 2
# Apply convolution
for i in range(offset, img_height - offset):
    for j in range(offset, img_width - offset):
        # Extract the region of interest (ROI)
        region = gray_img[i - offset:i + offset + 1, j - offset:j + offset + 1]

        # Apply the kernel to the ROI and get the result
        filtered_value = np.sum(region * kernel)
        
        # Assign the result to the output image, using clipping to handle overflow
        filtered_image[i, j] = np.clip(filtered_value, 0, 255)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(gray_img,cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Filtered Image')
plt.imshow(filtered_image, cmap='gray')
plt.show()

```

<h1>Power-Law Transform(Gamma Correction)

```python

# Load image in grayscale
img = cv2.imread('image/rango2.jpg', cv2.IMREAD_GRAYSCALE)

# Normalize the image to the range [0, 1]
img_normalized = img / 255.0

# Power-law (gamma correction) transformation
gamma = 0.5  # Adjust gamma value here (gamma < 1 brightens, gamma > 1 darkens)
gamma_corrected = np.power(img_normalized, gamma)

# Convert back to 8-bit range (0-255)
gamma_corrected = np.uint8(gamma_corrected * 255)

# Display original and gamma corrected images
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1, 2, 2)
plt.title(f'Gamma Corrected (gamma={gamma})')
plt.imshow(gamma_corrected, cmap='gray')

plt.show()
```
