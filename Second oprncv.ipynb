```python
#create new image
black_img=np.zeros_like(img)
plt.imshow(black_img)
plt.axis('off')
```

```python
#drawing rectangle,circle,line,text
cv2.rectangle(black_img,pt1=(100,100),pt2=(20,250),color=(255,0,0),thickness=3)
plt.imshow(black_img)
plt.axis('off')
```

```python
cv2.circle(black_img,center=(100,100),radius=20 ,color=(0,255,0),thickness=1)
cv2.line(black_img,pt1=(100,100),pt2=(300,0),thickness=1,color=(0,200,200))
cv2.putText(black_img,text='Sifat',org=(50,50),fontScale=1,color=(255,255,255),thickness=1,lineType=cv2.LINE_AA,fontFace=cv2.FONT_ITALIC)
plt.imshow(black_img)
plt.axis('off')
```

```python
def draw(event,x,y,flags,params):
    print(f'point:({x},{y})')
    if event==1:
        cv2.circle(black_img,center=(x,y),radius=1,color=(0,0,255),thickness=-1)


cv2.namedWindow(winname='Window')
cv2.setMouseCallback('Window',draw)

while True:
    cv2.imshow("Window",black_img)
    if cv2.waitKey(1) & 0xFF ==ord('x'):
        break

cv2.destroyAllWindows()
```

```python


# Load a grayscale image
image = cv2.imread("image/map.jpg", cv2.IMREAD_GRAYSCALE)


blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)

# Apply Canny edge detector
edges = cv2.Canny(blurred_image, threshold1=50, threshold2=150)

# Display the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Detected Edges (Canny)")
plt.imshow(edges, cmap='gray')
plt.show()
```
