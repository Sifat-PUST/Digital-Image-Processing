```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('image/rango2.jpg')
```

```python

# একটি কালো ইমেজ তৈরি করা
height, width = 300, 300
black_img = np.zeros((height, width, 3), dtype=np.uint8)

# মাউস ক্লিক এবং হোভার ইভেন্ট হ্যান্ডলার
def get_position(event, x, y, flags, param):
    global black_img  # কালো ইমেজকে গ্লোবাল ভ্যারিয়েবল হিসেবে চিহ্নিত করা

    # ক্লিক ইভেন্ট (বাম ক্লিক)
    if event == cv2.EVENT_LBUTTONDOWN:
        print(f"Clicked position (x, y): ({x}, {y})")
    
    # হোভার ইভেন্ট (মাউস মুভ)
    elif event == cv2.EVENT_MOUSEMOVE:
        # কালো ইমেজের কপি তৈরি
        img_copy = black_img.copy()
        
        # পজিশন টেক্সট Overlay করা
        text = f"Position: ({x}, {y})"
        cv2.putText(img_copy, text, (x + 10, y + 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)
        
        # ইমেজ প্রদর্শন করা
        cv2.imshow("Image", img_copy)

# উইন্ডো তৈরি এবং মাউস ক্লিক ইভেন্ট সেট করা
cv2.imshow("Image", black_img)
cv2.setMouseCallback("Image", get_position)

cv2.waitKey(0)  # যেকোনো কী চাপলে প্রোগ্রাম বন্ধ হবে
cv2.destroyAllWindows()
```

```python
def get_position(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:  # বাম ক্লিক করলে
        print(f"Clicked position (x, y): ({x}, {y})")
# উইন্ডো তৈরি এবং মাউস ক্লিক ইভেন্ট সেট করা
cv2.imshow("Image", img)
cv2.setMouseCallback("Image", get_position)

cv2.waitKey(0)  # যেকোনো কী চাপলে প্রোগ্রাম বন্ধ হবে
cv2.destroyAllWindows()
```

```python
def draw(event,x,y,flags,params):
    print(f'point:({x},{y})')
    if event==1:
        cv2.circle(black_img,center=(x,y),radius=1,color=(0,0,255),thickness=-1)


cv2.namedWindow(winname='Window')
cv2.setMouseCallback('Window',draw)

while True:
    cv2.imshow("Window",black_img)
    if cv2.waitKey(1) & 0xFF ==ord('x'):
        break

cv2.destroyAllWindows()
```
